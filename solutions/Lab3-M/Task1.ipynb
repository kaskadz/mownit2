```julia
using DataFrames
using Plots
using Statistics
using CSV
```

```julia
function naive_multiplication(A, B)
    C = zeros(Float64, size(A, 1), size(B, 2))
    for i = 1:size(A, 1)
        for j = 1:size(B, 2)
            for k = 1:size(A, 2)
                C[i, j] = C[i, j] + A[i, k] * B[k, j]
            end
        end
    end
    C
end
```

```julia
function better_multiplication(A, B)
    C = zeros(Float64, size(A, 1), size(B, 2))        
    for i = 1:size(B, 2)                           
        for j = 1:size(A, 2)                       
            for k = 1:size(A, 1)                   
                C[k, i] = C[k, i] + A[k, j] * B[j, i]
            end
        end
    end
    C
end
```

```julia
function run_naive_multiplication(x1, y1, x2, y2)
    A = rand(Float64, x1 * y1 )
    A = reshape(A, y1, x1)

    B = rand(Float64, x2 * y2 )
    B = reshape(B, y2, x2)

    return @elapsed naive_multiplication(A, B)
end
```

```julia
function run_better_multiplication(x1, y1, x2, y2)
    A = rand(Float64, x1 * y1 )
    A = reshape(A, y1, x1)

    B = rand(Float64, x2 * y2 )
    B = reshape(B, y2, x2)

    return @elapsed better_multiplication(A, B)
end
```

```julia
function run_lib_multiplication(x1, y1, x2, y2)
    A = rand(Float64, x1 * y1 )
    A = reshape(A, y1, x1)

    B = rand(Float64, x2 * y2 )
    B = reshape(B, y2, x2)

    return @elapsed A * B
end
```

```julia
sizes = [1, 50, 100, 500, 1000, 1100, 1200, 1300, 1400, 1500]

measurements = size(sizes)[:1]
repetitions = 10
data_count = measurements * repetitions

println("measurements:")
println(measurements)
println("repetitions:")
println(repetitions)
```

```julia
df = DataFrame(Size = Int64[], Type = String[], Time = Float64[])
show(df)
```

```julia
for i = 1:measurements
    for j = 1:repetitions
        time = run_naive_multiplication(sizes[i], sizes[i], sizes[i], sizes[i])
        
        push!(df, (sizes[i], "naive", time))

        print((i - 1) * measurements + j)
        print("/")
        println(data_count)
    end
end
```

```julia
for i = 1:measurements
    for j = 1:repetitions
        time = run_better_multiplication(sizes[i], sizes[i], sizes[i], sizes[i])
        
        push!(df, (sizes[i], "better", time))

        print((i - 1) * measurements + j)
        print("/")
        println(data_count)
    end
end
```

```julia
for i = 1:measurements
    for j = 1:repetitions
        time = run_lib_multiplication(sizes[i], sizes[i], sizes[i], sizes[i])
        
        push!(df, (sizes[i], "lib", time))

        print((i - 1) * measurements + j)
        print("/")
        println(data_count)
    end
end
```

```julia
df_agg = by(df, [:Size, :Type]) do subGroup
    DataFrame(AvgTime = mean(subGroup.Time), StDev = std(subGroup.Time))
end
```

```julia
sort(df_agg, [:Size, :AvgTime])
```

```julia
df_agg[:Type] .== "naive"
df_naive = df_agg[df_agg[:Type] .== "naive", :]
```

```julia
df_better = df_agg[df_agg[:Type] .== "better", :]
```

```julia
df_lib = df_agg[df_agg[:Type] .== "lib", :]
```

```julia
plot(df_lib[:Size], [df_naive[:AvgTime], df_better[:AvgTime], df_lib[:AvgTime]], label = ["naive", "better", "lib"], size = (1000, 600))
scatter!(df_lib[:Size], df_naive[:AvgTime], yerr = df_naive[:StDev], label = "naive")
scatter!(df_lib[:Size], df_better[:AvgTime], yerr = df_better[:StDev], label = "better")
scatter!(df_lib[:Size], df_lib[:AvgTime], yerr = df_lib[:StDev], label = "lib")
```
